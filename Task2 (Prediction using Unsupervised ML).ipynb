### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

### Load dataset

```python
Iris=load_iris()
X=Iris.data
y=Iris.target
```

### Range of clusters

```python
num_clusters= range(1,10)
inertia=[]
for n in num_clusters:
    kmeans=KMeans(n_clusters=n, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

### Plot

```python
plt.figure(figsize=(8, 6))
plt.plot(num_clusters, inertia, marker='o', color='b')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

### Fit KMeans with the optimal number of clusters

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X)
```

##### centroids

```python
centroids = kmeans.cluster_centers_
```

##### Plot data points for each cluster

```python
for i in range(optimal_k):
     # Select data points associated with the current cluster
    cluster_points = X[kmeans.labels_ == i]
    
    # Scatter plot for the current cluster
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i+1}')
```

##### Add centroids to the plot

```python
for i in range(optimal_k):
     # Select data points associated with the current cluster
    cluster_points = X[kmeans.labels_ == i]
    
    # Scatter plot for the current cluster
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i+1}')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=100, c='red', label='Centroids')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clusters and Centroids')
plt.legend()
plt.show()
```
